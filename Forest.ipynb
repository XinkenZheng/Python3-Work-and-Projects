```python
import re
class Forest():

    def __init__ (self, ntree = [3,7,1,9], widen=1, margin=0, bg=Tree.SPACE):
        """Constructor: build trees of layers specified in the list ntree. (The default is 4 trees of 3-, 7-, 1-, 9- layers.)
            * self.lines is a list of strings use to draw a forest in ASCII
        """
        self.ntree = ntree
        self.widen = widen
        self.margin = margin
        self.bg = bg
        self.Forest = []
        self.trees = []
        for i in range(len(ntree)):
            self.Forest.append(Tree(ntree[i],widen,margin,bg))
        self.r = False

        
    def findMax(self):
        length = 0
        max1 = 0 
        for i, tree in enumerate(self.Forest):
            if len(tree) > length:
                length = len(tree)
                max1 = i
        return max1
    
    def treeLines(self):
        treeLines = []
        max1 = self.findMax()
        for i, tree in enumerate(self.Forest):
            treeLines.append(tree.getLines())
        for i, tree in enumerate(treeLines):
            if len(tree) < len(treeLines[max1]):
                for j in range(len(treeLines[max1])-len(tree)):
                    tree.insert(0,(len(tree[-1]))*" ")
        return treeLines
    
    def setLines(self): # you may or may not implement this funcition
        """Compose a list of strings that represents the forest by combining the strings that 
            represent each individual tree.
           This is a helper function for build and rotate """
        treeLines = self.treeLines()
        trees = []
        
        
        for i in range(len(treeLines[-1])):
            line =""
            for tree in treeLines:
                line += tree[i]
            trees.append(line)
        self.trees = trees
    
    def getForest(self):
        return self.trees
    
    def getTrees(self):
        trees = []
        treeLines = self.treeLines()
        
        
        for i in range (len(treeLines)):
            trees.append(list(treeLines[i]))
        return trees
        
    # Note: Change the default of bg to None   
    def draw(self, reverse=False, bg=None, margin = 0, border=0 ):
        """Print out the list of strings that represents the forest with
            * background symbol=bg --> default was changed to None
            * the number of spaces to the left boarder of the window = margin (default 0)
           If reverse=True, the forest will be printed out upside down 
           If border>0, then the border of the objects will be replaced with the special symbols. 
        """
        for i in self.Forest:
            i.addBorder(i.getLines(),bg,border)
        if self.r == False:
            self.setLines()
        tree = self.trees
        if reverse:
            self.trees.reverse()
        for i in range(len(self.trees)):
            if bg != None:
                if bg != " ":
                    tree[i] = tree[i].replace(' ',bg)   
            print(self.trees[i]) 
        
    def  rotate (self, individual_rotation=None):
        """Revise the list of strings self.lines that represents the rotated trees.
            if individual_rotation=k is specified, then tree k is rotated 90 degree clockwise individually.
            If individual_rotation=None, then the entire forest will be rotated 90 degree clockwise.
        """
        if individual_rotation != None:
            self.Forest[individual_rotation].rotate()
       
        else:
            self.r = True
            self.setLines()
            n = len(self.trees[0])
            ans = []
            for i in range(n):
                x = ""
                for s in self.trees[::-1]:
                    x += s[i]
                ans.append(x)
            self.trees = ans
            
```
