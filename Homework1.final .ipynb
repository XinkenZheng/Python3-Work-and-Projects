```python
import re
class Tree:
    SPACE = " "
    FG = "*"
    NEWLINE = "\n"
    def __init__(self, layers=7, widen = 1, margin=0, bg= SPACE):
        self.lines = []
        self.bg = bg
        self.layers = layers 
        self.widen = widen
        self.margin = margin
        self.build (widen, bg)
        
    def line(self, margin, numChar, fg=FG): 
        newLine = (fg* numChar).center(margin)
        return newLine
    
    def triangle (self, margin, half_width, fg=FG):
        s = []
        hw = half_width*self.widen  
        bot_width = (2*hw) + 1
        height = ((hw)+1) //2 +1 
        if self.widen == 1:
            tp_width = (bot_width) - 2 * height +2
        else: 
            tp_width = (bot_width) - 2 * height
        for i in range(hw+1):
            l = self.line(margin, (2*i)+1,fg)
            total = 2*i +1
            if total >= tp_width and total <= bot_width:
                s.append(l)
        return s

    def treeTop (self, margin,n):
        tree = []
        counter = 0
        for i in range(n+1):
            y = self.triangle(margin, i,str(counter))
            counter +=1
            tree = tree + y
        return tree
    
    ###########################################
    def treeTrunk (self, margin,layers):
        width = ((self.layers*self.widen)//3)*2+1
        height = 2*width
        trunk = []
        for i in range(height):
            l = self.line(margin,width,fg = 'T')
            trunk.append(l)
        return trunk

    def build (self, widen=1, bg=SPACE):
        self.bg = bg
        myTreeTop = self.treeTop(self.margin,self.layers)
        myTreeTrunk = self.treeTrunk(self.margin,self.layers)
        self.lines = myTreeTop + myTreeTrunk
        maxLength = 1
        for length in range(len(self.lines)):
                if len(self.lines[length]) > maxLength:
                    maxLength = len(self.lines[length])
        for item in range(len(self.lines)):
                if len(self.lines[item]) <  maxLength:
                    symbols = (maxLength - len(self.lines[item]))//2
                    bg_symbols = self.bg * symbols
                    self.lines[item] = self.bg * symbols +self.lines[item] + self.bg * symbols
        return self.lines


    
    def __len__(self):
        length = 0
        for item in self.lines:
            for line in item:
                length+=1
        return length
    
    def getLines (self):
        return self.lines
    
    def rotate (self,degrees):
        maxLength = 1
        for length in range(len(self.lines)):
            if len(self.lines[length]) > maxLength:
                maxLength = len(self.lines[length])
        if degrees == 90:
            m = self.lines[:]
            m.reverse()
            for i in range(maxLength): 
                for x in m: 
                    print(x[i], end =' ') 
                print()
        if degrees == 180:
            reverse = self.lines[:]
            reverse.reverse()
            for a in reverse: 
                print(a)
        if degrees == 270:
            for i in range(maxLength): 
                for x in self.lines: 
                    print(x[i], end =' ') 
                print()
        if degrees == 360:
            for items in self.lines:
                print(items)
        
    def draw (self, reverse=False, bg=SPACE, margin=0):        
        self.build(bg = bg)
        maxLength = self.layers*2 +1
        if reverse == True:
            self.lines.reverse()
            for item in self.lines: 
                    print(self.SPACE * margin +item)
        else:
            for line in self.lines:
                    print (self.SPACE * margin +line )
                    
```
